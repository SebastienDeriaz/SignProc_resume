```python
import numpy as np
from numpy import sin, cos, pi
import matplotlib.pyplot as plt
```

```python
N = 200
a = np.linspace(0, 2*np.pi, N)

Noise = 0.2
alpha = 1.2
beta = 2.1
X = (alpha*np.cos(a) + 0.3*np.sin(a) + np.random.random(a.shape)*Noise).reshape(-1,1)
Y = (beta*np.sin(a) + 0.5*np.cos(a) + np.random.random(a.shape)*Noise).reshape(-1,1)

print(X)

plt.plot(X, Y, '.')
plt.grid()
```

$$A=[x_i^2,y_i^2,x_i,y_i,x_iy_i]$$

$$ax^2+by^2+cx+dy+exy=1$$

$$\Phi=(A^T A)^{-1}A^T b=A^{+}b$$

```python
A = np.asmatrix(np.concatenate([X**2, Y**2, X, Y, X*Y], axis=1))

phi = (A.T @ A).I * A.T * np.ones_like(X)

print(phi)
```
